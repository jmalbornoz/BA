# 0.- Imports

```python
import matplotlib.pyplot as plt
from matplotlib import style
from matplotlib import cm
style.use('fivethirtyeight')
%matplotlib inline

from wordcloud import WordCloud
```

# 1.- Read data

```python
with open('data/reviews.txt', 'r') as f:
    reviews = f.readlines()
```

```python
num_reviews = len(reviews)
```

# 2.- Cleanse data

```python
for k in range(num_reviews):
    reviews[k] = reviews[k].replace("'\n", '').replace("b'", '')
```

# 3.- Generate a single text file with all reviews

```python
text = ''
for k in range(num_reviews):
    text = text + reviews[k].lower()
```

# 4.- Remove words that do not provide insight

```python
text = text.replace('ba', '').replace('flight', '').replace('british', '').replace('airways', '').replace('london', '') \
           .replace('airline', '').replace('thi', '').replace('wa', '').replace(' s ', '')
```

# 5.- Generate Word Cloud

```python
word_cloud = WordCloud(width = 800, height = 800, collocations = False, background_color = 'black').generate(text)
```

```python
plt.figure(figsize=[8,10])
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('word_cloud.png');
plt.show();
```
