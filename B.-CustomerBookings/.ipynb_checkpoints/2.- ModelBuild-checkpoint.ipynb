# 0.- Imports

```python
import pandas as pd
import numpy as np
import time

import matplotlib.pyplot as plt
from matplotlib import style
from matplotlib import cm
style.use('fivethirtyeight')
%matplotlib inline
plt.rcParams["figure.figsize"] = (8,6)

import seaborn as sns

from sklearn.metrics import confusion_matrix, roc_curve, auc, roc_auc_score
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold, GridSearchCV
from sklearn.model_selection import cross_val_score

from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

# maximum number of dataframe rows and columns displayed
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 50)

RANDOM_STATE = 801
pd.options.mode.chained_assignment = None
```

# 1.- Read data

```python
df = pd.read_csv("data/customer_booking.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.dtypes
```

# 2.- Feature engineering

## 2.1.- Day of booking

We will map days of the week into numbers.

```python
df["flight_day"].unique()
```

```python
mapping = {
    "Mon": 1,
    "Tue": 2,
    "Wed": 3,
    "Thu": 4,
    "Fri": 5,
    "Sat": 6,
    "Sun": 7,
}

df["flight_day"] = df["flight_day"].map(mapping)
```

```python
df["flight_day"].unique()
```

## 2.2.- Feature engineering pipeline

We begin by splitting the dataset into the target and the (hopefully) predictive features.

```python
target_name = "booking_complete"
features, target = df.drop(columns=target_name), df[target_name]
```

We identify numerical and categorical features. 

```python
from sklearn.compose import make_column_selector as selector

numerical_columns_selector = selector(dtype_include=np.number)
categorical_columns_selector = selector(dtype_include=object)

numerical_columns = numerical_columns_selector(features)
categorical_columns = categorical_columns_selector(features)
```

```python
numerical_columns
```

```python
categorical_columns
```

```python
df[categorical_columns].nunique()
```

Given that two of our categorical features have high cardinality, we will use an ordinal encoder in our preprocessing pipeline. No numerical preprocessing is required as the features have no missing values; besides, the classifiers that we will build do not require feature scaling.

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer

# as there are categories that may not be present in the test data, we will set the 'handle_unknown'
# parameter for the ordinal encoder
categorical_preprocessor = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)

# no preprocessing required for numerical variables 
preprocessor = ColumnTransformer(
    [('cat_preprocessor', categorical_preprocessor, categorical_columns)],
    remainder='passthrough', sparse_threshold=0)
```

## 2.3.- Train/test split

We will use a stratified train/test split since we have an imbalanced target.

```python
data_train, data_test, target_train, target_test = \
            train_test_split(features, target, stratify=target, random_state=RANDOM_STATE)
```

# 3.- Predictive model

We will train two baseline models: a random forest classifier and an XGBoost classifier.

```python
# we make sure that the proportion of negative/positive class remains the same across the CV folds
cv = StratifiedKFold(n_splits=5)
```

## 3.1.- Random forest baseline model

```python
# random forest 
model_rf = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced'))
])
```

```python
# evaluate model
scores_rf = cross_val_score(model_rf, data_train, target_train, cv=cv, scoring='roc_auc', n_jobs=-1)
```

```python
scores_rf
```

```python
# summarize performance, random forest model
print('Mean AUC score Random Forest: %.5f' % scores_rf.mean())
```

We now train the model on all available training data, and analyse the model's performance through using the confusion matrix and ROC curve.

```python
model_rf.fit(data_train, target_train)
```

```python
predictions_rf = model_rf.predict(data_test)
```

### 3.1.1.- Random forest model: confusion matrix

```python
confusion_rf = confusion_matrix(target_test, predictions_rf)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.heatmap(confusion_rf, annot=True, fmt="d", linewidths=.5)
ax.set(xlabel='Predicted', ylabel='Actual')
ax.set_title('Baseline Random Forest Model')
ax.set_xticklabels(['Not completed', 'Completed'])
ax.set_yticklabels(['Not completed', 'completed']);
```

### 3.1.2.- Random forest model: ROC curve

```python
probs = model_rf.predict_proba(data_test)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(target_test, preds)
roc_auc = auc(fpr, tpr)
```

# 4.- XGBoost hyperparameter tuning

We will now tune the hyperparameters of our XGBoost classifer using the bayesian approach available in the `hyperopt` library. 

We begin by aplying the preprocessing pipeline to the train and test sets. 

```python
x_train = preprocessor.fit_transform(data_train)
```

```python
x_test = preprocessor.transform(data_test)
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
```

We define our parameter space.

```python
space = {'learning_rate' : hp.uniform("learning_rate", 0.01, 1),
         'max_depth': hp.randint("max_depth", 1, 11),
         'gamma': hp.uniform ('gamma', 1, 9),
         'reg_alpha' : hp.uniform('reg_alpha', 0, 10),
         'reg_lambda' : hp.uniform('reg_lambda', 0, 10),
         'colsample_bytree' : hp.uniform('colsample_bytree', 0.5, 1),
         'min_child_weight' : hp.uniform('min_child_weight', 0, 10),
         "n_estimators": hp.randint("n_estimators", 3, 10000),
}
```

And our objective function.

```python
def hyperparameter_tuning(space):
    
    model = XGBClassifier(learning_rate=space['learning_rate'], max_depth=space['max_depth'],
                          gamma=space['gamma'], reg_alpha = space['reg_alpha'],
                          reg_lambda = space['reg_lambda'], colsample_bytree= space['colsample_bytree'], 
                          min_child_weight=space['min_child_weight'], n_estimators=space['n_estimators'], 
                          random_state=RANDOM_STATE, scale_pos_weight=weights, use_label_encoder=False)
    
    evaluation = [(x_train, target_train), (x_test, target_test)]
    
    model.fit(x_train, target_train, eval_set=evaluation, eval_metric=['auc','error','logloss'], verbose=False)

    auc = roc_auc_score(target_test, model.predict_proba(x_test)[:, 1])
    print ("SCORE:", auc)
    
    #change the metric if you like
    return {'loss': -auc, 'status': STATUS_OK, 'model': model}
```

```python
start_time = time.time()
trials = Trials()
best = fmin(fn=hyperparameter_tuning, space=space, algo=tpe.suggest, max_evals=100, trials=trials)
end_time = time.time()
print(f"Hyperparameter optimisation took {(end_time - start_time)/60} minutes to run.")

print(best)
```

### 4.1.4.- Decision threshold setting: payoff matrix

In order to select a decision threshold for our bookings predictive model we will use a payoff matrix in order to identify the threshold value that maximises profit. 

We will assume that the payoff matrix for this business use case corresponds to the following outcomes:

* a true positive means a completed booking has been correctly predicted, with a profit of ¬£200
* a true negative means an incomplete booking has been correctly predicted. In this case the cost of nudging the customer will be -¬£5
* a false positive means that an incomplete booking was incorrectly predicted as completed. In this case there is an administrative cost of -¬£50
* a false negative means that a completed booking was incorrected predicted as incomplete, meaning administrative costs for the company. In this case there is a loss of -¬£50.

We will represent this scenario with the payoff matrix below:

```python
payoff = np.array([[-5, -50], [-50, 200]])
```

With this information we can now plot the average payoff for each decision threshold. We are interested in the optimal decision threshold that maximises average payoff.

```python
avg_payoffs = []

max_payoff = -80000
optimal_threshold = -1
for threshold in thresholds:
    cm = confusion_matrix(target_test, probs[:,1] > threshold)
    avg_payoffs += [np.sum(cm * payoff) / np.sum(cm)]
    if avg_payoffs[-1] > max_payoff:
        max_payoff = avg_payoffs[-1]
        optimal_threshold = threshold
```

```python
max_payoff
```

```python
optimal_threshold
```

```python
plt.plot(thresholds, avg_payoffs)
plt.title('Effect of threshold on average payoff')
plt.axvline(optimal_threshold, color='orange', linestyle='dotted', label='Optimal threshold')
plt.xlabel('Threshold')
plt.ylabel('Average payoff')
plt.legend()
```

With this optimal threshold we will now compute the corresponding precision and recall values for our model.

```python
from sklearn.metrics import precision_score, recall_score

prediction_optimal = probs[:,1] > optimal_threshold

print(f'The precision score that corresponds to maximum payoff is {precision_score(target_test, prediction_optimal)}')
print(f'The recall score that corresponds to maximum payoff is {recall_score(target_test, prediction_optimal)}')
```

We finish with the confusion matrix for the optimal decision threshold.

```python
confusion_xgb_tuned_optimal = confusion_matrix(target_test, prediction_optimal)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.heatmap(confusion_xgb_tuned_optimal, annot=True, fmt="d", linewidths=.5)
ax.set(xlabel='Predicted', ylabel='Actual')
ax.set_title('Tuned XGBoost Model, optimal threshold')
ax.set_xticklabels(['Not completed', 'Completed'])
ax.set_yticklabels(['Not completed', 'completed']);
```

### 4.1.5.- Feature importance: permutation importance

```python
from sklearn.inspection import permutation_importance

r = permutation_importance(model_xgb_tuned, data_test, target_test, n_repeats=30, random_state=RANDOM_STATE)
```

```python
r_sub = dict((k, r[k]) for k in ('importances_mean', 'importances_std'))
permutation_importances_df = pd.DataFrame(r_sub)
permutation_importances_df['Features'] = data_test.columns
permutation_importances_df['importance_norm'] = permutation_importances_df['importances_mean']\
                                               /permutation_importances_df['importances_mean'].sum()
permutation_importances_df.sort_values(by='importance_norm', ascending=False, inplace=True)
permutation_importances_df
```

```python
plt.rcParams["figure.figsize"] = (12,8)
ax1 = permutation_importances_df.iloc[0:12].sort_values(by=['importance_norm']).plot.barh(x='Features', 
                                                                                          y = 'importances_mean')
plt.title('Permutation importance', fontsize=20)
plt.xlabel('Normalised importance', fontsize=15)
ax1.yaxis.set_label_position("right")
plt.ylabel('Features', fontsize=16)
ax1.get_legend().remove()
plt.savefig('permutation_importance.png');
```
